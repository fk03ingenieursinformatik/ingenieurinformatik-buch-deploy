# Namensräume und Seiteneffekte

Jeder Name (Variable) in einem Programm hat einen Sichtbarkeitsbereich, eine Lebensdauer und gehört zu einem Namensraum.
Wenn Funktionen veränderliche Objekte erhalten und diese ändern, spricht man von Seiteneffekten.
Dieser Abschnitt erklärt die Grundlagen und hilft Ihnen, typische Fehler zu vermeiden.
