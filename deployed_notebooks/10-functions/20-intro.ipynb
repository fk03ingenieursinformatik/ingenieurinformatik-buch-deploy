# Definition und Aufruf (A)

In diesem Abschnitt lernen Sie, wie Sie eigene Funktionen definieren und aufrufen.
Dazu gehören die Syntax mit `def`, Parameter und Rückgabewerte, Standardargumente sowie praktische Anwendungsbeispiele.

Im Kapitel [Kontrollstrukturen](sec-functions-control-structures) haben Sie bereits die Funktion `create_bundle()` kennengelernt.
Hier geht es nun darum, solche Funktionen selbst zu schreiben und ihre Schnittstellen (Parameter, Rückgabewerte) gezielt zu gestalten.
