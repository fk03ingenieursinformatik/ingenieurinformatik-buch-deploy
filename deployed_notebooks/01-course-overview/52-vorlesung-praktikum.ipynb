Es kann deshalb vorkommen, dass Inhalte, die im Praktikum benötigt werden, in der Vorlesung noch nicht explizit oder noch nicht vollständig behandelt wurden.
- Wir arbeiten aber auch in der Vorlesung früh mit Beispielen, sodass Sie im Praktikum nicht „ins kalte Wasser“ springen.
- Zu Beginn jedes Praktikums gibt es außerdem eine kurze Einführung und Einweisung. (Im Sinne unserer Metapher heißt das zum Beispiel: „Heute schalten wir vom 2. in den 3. Gang“, auch wenn wir die Details des „Getriebes“ erst später in der Vorlesung besprechen.)
