```python

    x = A0

```

als 

```python
x = A0 if P0 else A1
````

zu schreiben.

`````python
y = 12
x = 0 if y % 2 == 0 else 1
x
`````

Diese Schreibweise l√§sst sich gut mit sog. [Comprehension](sec-comprehension) verbinden:

`````python
numbers = list(range(10))
even = [True if x % 2 == 0 else False for x in numbers]
print(f'numbers: {numbers}')
print(f'even: {even}')
`````
